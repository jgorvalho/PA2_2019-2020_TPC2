# Exerc√≠cio 1


```python
x=df.month_number
y=df.total_profit
plt.plot(x,y)
plt.xticks(x, x, rotation='horizontal')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title("Company profit per month")
plt.show()
```

```python
sum_column = df.sum(axis=0)
TotalS = sum_column


TotalSalesFC = (TotalS.facecream/TotalS.total_units)
TotalSalesFW = (TotalS.facewash/TotalS.total_units)
TotalSalesTP = (TotalS.toothpaste/TotalS.total_units)
TotalSalesBS = (TotalS.bathingsoap/TotalS.total_units)
TotalSalesSH = (TotalS.shampoo/TotalS.total_units)
TotalSalesMO = (TotalS.moisturizer/TotalS.total_units)

data = [TotalSalesFC,TotalSalesFW,TotalSalesTP,TotalSalesBS,TotalSalesSH,TotalSalesMO]
label = ["Face Cream","Face Wash","Tooth Paste","Bathing Soap","Shampoo","Moisturizer"]

plt.axis("equal")
plt.pie(data,labels = label,radius=1.5,autopct = "%0.0f%%", startangle=300)


plt.show()
```
