## Import Libraries

```python
# libraries
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import seaborn as sns
sns.set_style("white")
import pandas as pd
my_dpi=96
```

## Get Data 

```python
url = 'https://python-graph-gallery.com/wp-content/uploads/gapminderData.csv'
data = pd.read_csv(url)
#data=pd.read_csv('gapminderData.csv')
data.head()
```

```python
data['continent']=pd.Categorical(data['continent'])
```

```python
anos=data.year.unique()
```

## Gráfico Animação

```python
%matplotlib notebook
%matplotlib notebook

# initialize 
#fig=plt.figure(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi)
fig, ax = plt.subplots(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi)

# Add titles
plt.yscale('log')
plt.xlabel("Life Expectancy")
plt.ylabel("GDP per Capita")
plt.ylim(100,100000)
plt.xlim(30, 90)

def animate(i):
    plt.title("Year: "+str(anos[i]) )
    tmp=data[ data.year == anos[i] ]
    scut=plt.scatter(tmp['lifeExp'], tmp['gdpPercap'] , s=tmp['pop']/200000 , c=tmp['continent'].cat.codes, cmap="Accent", alpha=0.5, edgecolors="white", linewidth=1)
    #scut=ax.scatter(tmp['lifeExp'], tmp['gdpPercap'] , s=tmp['pop']/200000 , c=tmp['continent'].cat.codes, cmap="Accent", alpha=0.5, edgecolors="white", linewidth=1)
    
ani = animation.FuncAnimation(fig, animate, frames=len(anos), interval=200, blit=False, repeat=False)
    
plt.show()
```

## Questão A

Apresenta as seguintes variáveis: a alteração da life expectancy e GDP per capita demonstrado pelo avançar dos anos na **função animate**. Esta função vai buscar os anos á variavel "anos" (demonstra todos os anos existentes - data.year.unique()) e completa a informação com todos os continentes demonstrando assim no gráfico. 
Esta versão vai sobrepondo as imagens criadas demonstrando a evolução de todos os continentes de 5 em 5 anos num só gráfico.
